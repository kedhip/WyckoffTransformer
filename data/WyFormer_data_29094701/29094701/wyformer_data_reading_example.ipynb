```python
import pandas as pd
from monty.json import MontyDecoder
import os

# Check if the file exists
file_path = "mp_20/WyckoffTransformer/CrySPR/CHGNet_fix/data.csv.gz"
if not os.path.exists(file_path):
    # List files in the directory to help debug the path
    print("File not found. Files in directory:")
    print(os.listdir(os.path.dirname(file_path)))
else:
    chgnet_data = pd.read_csv(file_path, index_col="material_id", dtype=str).map(MontyDecoder().decode)
```
